```python
import numpy as np
import numpy.random as rnd
import matplotlib.pyplot as plt
```

```python
nTotal = 1000
nGenuine = 300
nNoisy = nTotal - nGenuine
```

```python
dRayleighParam = 5
aGenuineSignals = rnd.rayleigh(dRayleighParam,nGenuine)
aGenuineSignals.mean()
```

```python
dRayleighParam*np.sqrt(np.pi/2.0)
```

```python
dSig = 1.0
aGenSigX = aGenuineSignals + rnd.normal(0,dSig,nGenuine)
aGenSigY = aGenuineSignals + rnd.normal(0,dSig,nGenuine)
plt.plot(aGenSigX, aGenSigY,'ro')
plt.show()
```

```python
from scipy.stats import rankdata
```

```python
aRankGenX = rankdata(aGenSigX)
aRankGenY = rankdata(aGenSigY)
plt.plot(aRankGenX, aRankGenY,'r.')
plt.show()
```

```python
corrGen = np.dot(aRankGenX/nGenuine, aRankGenY/nGenuine)/nGenuine
```

```python
corrGen
```

```python
iChisqParam = 4
aNoisySigX = rnd.chisquare(iChisqParam, nNoisy)
aNoisySigX.mean()
```

```python
dGeoParam = 1.0/iChisqParam
aNoisySigY = rnd.geometric(dGeoParam, nNoisy)
aNoisySigY.mean()
```

```python
plt.plot(aNoisySigX, aNoisySigY,'b.')
plt.plot(aGenSigX, aGenSigY,'r.')

plt.show()
```

```python
aSigX = np.concatenate([aNoisySigX,aGenSigX])
aSigY = np.concatenate([aNoisySigY,aGenSigY])
plt.plot(aSigX, aSigY,'k.')
plt.show()
```

```python
aRankX = rankdata(aSigX)
aRankY = rankdata(aSigY)
plt.plot(aRankX, aRankY,'k.')
plt.show()
```

```python
corr = np.dot(aRankX/nTotal, aRankY/nTotal)/nTotal
```

```python
corr
```
